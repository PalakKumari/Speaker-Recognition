```python
import json
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow.keras as keras
import matplotlib.pyplot as plt
import math
import librosa, librosa.display
```

```python
#training Data
metadata_train = pd.read_csv("D:/Data science/INTERNSHIP/metadata_train.txt", sep = ' ', header = None)
metadata_train.columns = ['true/false','location1','location2']
# metadata_train= pd.DataFrame(metadata_train)
y=metadata_train['location1'].unique()
y = pd.DataFrame(y)
```

```python
y.columns=['location1']
```

```python
data = {
    "labels" :[],
    "value":[]
}
```

```python
for i in y.index[:3000]:
    sample = y.loc[i,'location1']
    sample_1_path = 'D:/Data science/INTERNSHIP/wav/'+sample
    file = sample_1_path
    signal, sample_rate = librosa.load(file, sr=16000)
    data["labels"].append(sample[0:7]) 
    data["value"].append(signal.tolist())
```

```python
data1=pd.DataFrame(data)
```
